# Checking Stationary Data

```python
from statsmodels.tsa.stattools import adfuller
```

```python
plt.figure(figsize = (12,8))
sb.lineplot(x = New_York.index, y = "Temperature", data = New_York)
```

```python
def Adfuller_Test(Temp):
    Result = adfuller(Temp)
    Label = ["Adf Test Stats", "P-Value", "Lags Used", "Number of observation"]
    for result, label in zip(Result, Label):
        print(label, ":", result)
    if Result[1] < 0.05:
        print("Strong evidence against the null hypothesis and so The data is stationary")
    else:
        print("Weak evidence against the null hypothesis, and so the data is not stationary")
    
    
```

```python
Adfuller_Test(New_York["Temperature"])
```

We need plot of pacf and acf and differencing to get the order that will be passed to the ARIMMA model
acf gives us q
pacf ggives p
d is the shifting used in the data. ranges from 1 and 4. depending on the shifting number used.

(p,d,q)
