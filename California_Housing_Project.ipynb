```python
#  Creating A Transformation Class to Automate the add columns transformation
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Best Estimator amd TransformerMixin Works with only numerica Data

class Transformation(BaseEstimator, TransformerMixin):
    pop_index = 5
    room_index = 3
    household_index = 6
    bedroom_index = 4
    def __init__(self, add_bedroom_per_room = True):
        self.add_bedroom_per_room = add_bedroom_per_room
        
    def fit(self, X, Y = None):
        return self
    
    def transform(self, X, Y = None):
        room_per_household = X[:, Transformation.room_index]/(X[:, Transformation.household_index])
        population_per_household = X[:, Transformation.pop_index]/(X[:, Transformation.household_index])
        
        if self.add_bedroom_per_room:
            bedroom_per_room = X[:, Transformation.bedroom_index]/(X[:, Transformation.room_index])
            return np.c_[X, room_per_household, population_per_household, bedroom_per_room]
        
        else:
            return  np.c_[X, room_per_household, population_per_household]
            
# np.c_ convert DataFrame or Series object to an array note: we have to index values we want as if we want to index a list    
        
```

```python
# Now Lets Create a pipeline to actually fixed all these together

# pipeline Takes two elements in a tuple, Name of the Transformer amd the Transformer method all in a list
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
Numeric_Pipeline = Pipeline([("imputer", SimpleImputer(strategy = "median")),
                    ("Transformation", Transformation(add_bedroom_per_room=True)),
                    ("Scaler", StandardScaler())
                   ])
```

#  Writing a function to automate the result for other Linear regression algorithm

```python
def Model(model, x_train, y_train, x_test, y_test):
    X = model()
    X.fit(x_train, y_train)
    pred = X.predict(x_test)
    rmse = np.sqrt(mean_squared_error(pred, y_test))
    print(f"for {model} model:")
    print()
    print(f"Root mean square error is {rmse}")
    
```
