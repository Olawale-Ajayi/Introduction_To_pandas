```python
#   Combining and merging Data

# suppose we have 2 dataframes below
```

```python
df3 = pd.DataFrame({"key" : ["b", "c", "b", "c", "c", "b", "b", "c"],
                    "data1": range(8)})
df3
```

```python
df4 = pd.DataFrame({"key": ["b", "c", "d"],
                   "data2": range(3)})
df4
```

```python
pd.merge(df3,df4)
```

```python
# there is need to specify the columns to join
```

```python
pd.merge(df3,df4, on = "key")  # because the two Frames shares the same columns name "key"(default how to joint is "inner)")
```

```python
pd.merge(df3,df4, on = "key", how = "left")
```

```python
pd.merge(df3,df4, on = "key", how = "right")
```

```python
# if the merging columns share different columns name as demonstrated below


```

```python
df5 = pd.DataFrame({"Lkey" : ["a", "c", "a", "d", "d", "b", "b", "c"],
                    "data1": range(8)})
df5
```

```python
df6 = pd.DataFrame({"Rkey": ["b", "c", "d", "a", "e"],
                   "data2": range(5)})
df6
```

```python
pd.merge(df5,df6, left_on = "Lkey", right_on= "Rkey").sort_values(by="Lkey")  # default how is inner which used the combinations of values in the merged columns
```

```python
# we can pass the how argument to determine how the resulting merger will look like
```

```python
pd.merge(df5,df6, left_on = "Lkey", right_on= "Rkey", how= "outer")  # return the union of the two frames
```

```python
pd.merge(df5,df6, left_on = "Lkey", right_on= "Rkey", how = "left")  # used intersection from the left side of the frame
```

```python
pd.merge(df5,df6, left_on = "Lkey", right_on= "Rkey", how = "right")  #  used intersection from right side of the frame
```

```python
Pivot4 = pd.DataFrame({"key": ["Microsoft", "Apple", "Google", "Tecno"],
                  "Data": [3,45,23, 28],
                  "worth": [23,84,38,39],
                  "Net": [25,48,29,10]})
Pivot4
```

```python
pivot5 = pd.melt(Pivot4, "key")  # thats an easy Unpivoting
pivot5
```

```python
pivot5.pivot("key", "variable", "value")  #  pivot reverses what melt just did
```

```python
pivot5.pivot("key", "variable", "value").reset_index()  #  here is resettin the index back to the data
```

```python
# we can specify the subsey of a variable to unpivot
```

```python
pd.melt(Pivot4, id_vars= "key", value_vars = ["Data", "Net"])   # here we are picking data and net to unpivot sing "key
#as the main data columns"
```

```python
pd.melt(Pivot4, id_vars= "key", value_vars = ["Data", "Net"]).pivot("key", "variable")   #  we can pass two column in the pivot columnss or all of them
```

```python
pd.melt(Pivot4, id_vars= "key", value_vars = ["Data", "Net"]).pivot("key", "variable", "value")
```

```python
pd.melt(Pivot4, id_vars= "key", value_vars = ["Data", "Net"]).pivot("key", "variable", "value").reset_index()
```

```python
pd.melt(Pivot4, id_vars= "key", value_vars = ["Data", "Net"]).pivot("key", "variable", "value").reset_index().set_index("key")
```
