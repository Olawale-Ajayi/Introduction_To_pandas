```python
#Tick, Label and Legend  # ("setting default axis names and customising chat")
```

```python
fig2 = plt.figure(figsize=(8,6))

axis = fig2.add_subplot(1,1,1)
axis.plot(np.random.randn(1000).cumsum(), color = "r", alpha= 0.75, linestyle = "solid", label = "red" )
ticks = axis.set_xticks([0,250,500, 750,1000])  # changes the range of the x_axis
labels = axis.set_xticklabels(["Mini","small", "mid", "Big", "Large"])  # rename the range set
title = axis.set_title("Welcome to my Chart") # gives title to the chart
xlabel = axis.set_xlabel("Dimensions")
ylabel = axis.set_ylabel("Size", rotation = 35)

```

```python
fig2 = plt.figure(figsize=(8,6))
Chart = fig2.add_subplot(1,1,1)
Chart.plot(np.random.rand(200).cumsum(), color = "red", linestyle = "dashed", label = "Men")
Chart.plot(np.random.randn(200).cumsum(), color = "blue", linestyle = "solid", label = "Female")
Chart.plot(np.random.randn(200).cumsum(), color = "y", linestyle = "dotted", label = "Unisex")
Chart.plot(np.random.randn(200).cumsum(), color = "brown", linestyle = "dashed", label = "Children")
Chart.legend(loc="best")
properties ={"title": "Dressing Category in New York Fashion House",
        "xlabel": "Category",
        "ylabel": "Size",
        "xticks": range(0,220,10),
        "xticklabels": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "B", "Z", "C", "D", "E", "F", "G", "H", "I", "J", "Y"]
    }

Chart.set(**properties)

# we can set all the chart properties as thus as oppose to naming 
```
